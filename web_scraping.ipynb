```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://books.toscrape.com"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
## finding all book containers
books = soup.find_all('article', class_="product_pod")
```

```python
books
```

```python
## initialize list sto store this information
titles = []
prices = []
ratings = []
availability = []
```

```python
## extracting information for each book
for book in books:

    title = book.h3.a['title']
    titles.append(title)

    price = book.find('p', class_="price_color").text
    prices.append(price)

    rating_class = book.find('p', class_="star-rating")['class'][1]
    ratings.append(rating_class)

    avail = book.find('p', class_="instock availability").text.strip()
    availability.append(avail)
```

```python
books_dataframe = pd.DataFrame({
    "Title": titles,
    "Price":prices,
    "Rating":ratings,
    "Availability":availability
})
```
